```python
import geopandas as gpd
import pandas as pd
import shapely
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
shapeFilePath = 'yelpNightlife/yelpBusiness.shp'
```

```python
yelpBarsDF = gpd.GeoDataFrame.from_file(shapeFilePath)
```

```python
yelpBarsDF.head()
```

```python
yelpBarsDF = yelpBarsDF[yelpBarsDF.geometry.apply(lambda x : x.coords[0][1]>34)]
```

```python
yelpBarsDF.plot()
```

```python
barLocDF = pd.read_csv('partynyc/bar_locations.csv')
```

```python
barLocDF.head()
```

```python
# converting latitude longitude into geometry POINT
barLocDF['lonlat'] = list(zip(barLocDF["Longitude"].values,barLocDF["Latitude"].values))
barLocDF['geometry'] = barLocDF[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
```

```python
crs = {'init': 'epsg:4326'}
```

```python
barLocGDF = gpd.GeoDataFrame(barLocDF, crs = crs)
```

```python
barLocGDF.plot()
```

```python
len(yelpBarsDF)
```

```python
len(barLocGDF)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
barLocGDF.plot(ax=ax, markersize=12, color = 'green')
yelpBarsDF.plot(ax=ax, markersize=7, color = 'red')
```

```python
from geopandas.tools import sjoin

tmp = gpd.sjoin(barLocGDF, yelpBarsDF, how='inner')
```

```python
len(tmp)
```

```python
barCircleGDF = barLocGDF.copy()
```

```python
barCircleGDF.geometry = barCircleGDF.geometry.apply(lambda x: x.buffer(0.0001))
barCircleGDF.head(2)
```

```python
tmp = gpd.sjoin(barCircleGDF, yelpBarsDF, how='inner')
len(tmp)
```

```python
len(set(tmp.lonlat))
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
tmp.plot(ax=ax, markersize=10, color = 'blue')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
barCircleGDF.plot(ax=ax, markersize=5, color = 'green')
yelpBarsDF.plot(ax=ax, markersize=5, color = 'red')
```
